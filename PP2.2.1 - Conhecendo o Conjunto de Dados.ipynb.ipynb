## Lista dos tipos de solo:

```python
soil_cols = [col for col in df.columns if 'Soil_Type' in col]
```

## Lista das áreas de natureza selvagem:

```python
area_cols = [col for col in df.columns if 'Wilderness_Area' in col]
```

## Junção das duas listas:

```python
lista = [*soil_cols, *area_cols]
```

## Lista dos atributos preditores:

```python
soil_type = 'Soil_Type'
wilderness_area = 'Wilderness_Area'
```

```python
preditores = [col for col in df.columns[:-1] if col not in lista]
preditores += [soil_type, wilderness_area]
```

## Atributo-alvo:

```python
alvo = 'Cover_Type'
```

## Criação das novas colunas:

```python
df[soil_type] = df[soil_cols].idxmax(axis=1).str.replace(soil_type, '').astype('category')
```

```python
df[wilderness_area] = df[area_cols].idxmax(axis=1).str.replace(wilderness_area, '').astype('category')
```

## Visualização da distribuição, tendência central e dispersão:

```python
for coluna in preditores:
    plt.figure(figsize=(12, 5))
    sns.histplot(df[coluna], kde=True)
    plt.title(f'Distribuição de {coluna}')
    plt.xlabel(coluna)
    plt.ylabel('Frequência')
    plt.tight_layout()
    plt.show()
    print()
```

```python
df[preditores].describe(include='all').T
```

## Frequência das classes:

```python
df[alvo].value_counts(normalize=True) * 100
```

```python
sns.countplot(x=alvo, data=df)
plt.title(f'Distribuição do atributo-alvo ({alvo})')
plt.show()
```
